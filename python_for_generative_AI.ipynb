```python
#while
i = 0 #initialization
while i<5:
  print(i)
  i+=1
```

```python
#break--this statement break the loop
i = 0
while i<5:
  i+=1
  if i == 3:
    break
  else:
    print(i)
```

```python
#continue statement--skip the current iteration
i = 0
while i<5:
  i+=1
  if i == 3:
    continue
    print(i)
  else:
    print(i)
```

```python
#pass statement---pass the code
i = 0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```

```python
#bar plot--it is work on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 90, 80])
color = np.array(["red", "yellow", "orange", "green"])
plt.bar(fruit, quantity, color = color)
plt.show()
```

```python
fruit = np.array(["Apple", "Banana", "Orange", "Watermelon"])
quantity = np.array([100, 120, 90, 80])
color = np.array(["red", "yellow", "orange", "green"])
plt.barh(fruit, quantity, color = color)
plt.show()
```

```python
#pie plot
fruit_percentage = np.array([20, 25, 35, 20])
fruit_name = np.array(["Apple", "Banana", "Orange", "Watermelon"])
colors = np.array(["red", "yellow", "orange", "green"])
myexplode = np.array([0.2, 0, 0, 0])
plt.pie(fruit_percentage, labels = fruit_name, autopct="%1.1f%%",
        colors =colors, explode = myexplode  )
plt.legend(title = "Fruit", loc = "best")
plt.show()
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

# compute cosine similarity
cosine = np.dot(A,B)/(norm(A) * norm(B))
print("Cosine Similarirty:",cosine)

```

```python
from transformers import pipeline
summarizer = pipeline ("summarization" , model = "facebook/bart-large-cnn" )

text = """The most common name for rhododendron in India is Burans (or variations like Burans, Bras, or Buras).
Other common names vary by region, including Lal buransh (Hindi), Billi (Tamil), Kattupoo varasu (Malayalam), and Pu (Kannada)."""

summary =summarizer(text,max_length=30,min_length=10, do_sample=False)
print(summary[0])
```
